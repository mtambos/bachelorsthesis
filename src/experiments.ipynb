```python
%%px --noblock
from __future__ import division, print_function
from cStringIO import StringIO
from datetime import datetime, time, timedelta
from collections import OrderedDict
import csv
from functools import partial
import os
import shutil
import sys

os.chdir('/home/mtambos/ba_src')

import bottleneck as bn
from dateutil.parser import parse as du_parse
import numpy as np
import Oger as og
import pandas as pd
import seaborn as sns
from scipy.stats import norm

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib as mpl
mpl.rc("figure", figsize=(16, 8))

import plotly.plotly as py
import plotly.tools as tls
from plotly.graph_objs import *
py.sign_in("mtambos", "4zd8yv43y9")

import utils
import signal_generation
#from nupic.support.configuration import Configuration
#Configuration.readConfigFile('nupic-default.xml', path='/home/mtambos/nupic/config/default')
from cla.cla import open_input_file, prepare_run, process_row, create_model
from cla.swarm import swarm

from mgng.amgng import AMGNG, compare_models
from mgng.mgng import distances, find_winner_neurons

class Capturing(list):
    def __enter__(self):
        self._stdout = sys.stdout
        sys.stdout = self._stringio = StringIO()
        return self
    def __exit__(self, *args):
        self.extend(self._stringio.getvalue().splitlines())
        sys.stdout = self._stdout


def plot_signal(signal, threshold, density_column, regex,
                annotation_column, increase_start, decrease_stop,
                slce=None, use_plotly=False, annotate_scores=True,
                plot_score=True, plot_threshold=True):
    if slce is not None:
        signal = signal[slce]
    f1_score = utils.f1_score(threshold, signal, density_column,
                              regex, annotation_column=annotation_column)
    f = plt.figure()
    nro_subplots = 2
    if plot_score:
        nro_subplots = 3
    ax1 = plt.subplot(nro_subplots,1,1)
    for c in signal.columns:
        if 'signal' in c:
            signal[c].plot(ax=ax1, alpha=.5)
    ax1.axvspan(signal.index[increase_start],
                signal.index[decrease_stop],
                alpha=.5)
    ax2 = plt.subplot(nro_subplots,1,2)
    signal[density_column].plot(ax=ax2)
    if plot_threshold:
        ax2.axhline(threshold)
    ax2.axvspan(signal.index[increase_start],
                signal.index[decrease_stop],
                alpha=.5)
    if plot_score:
        ax3 = plt.subplot(nro_subplots,1,3)
        signal.anomaly_score.plot(ax=ax3)
        ax3.axvspan(signal.index[increase_start],
                    signal.index[decrease_stop],
                    alpha=.5)
        ax3.set_ylabel('anomaly_score')
        ax3.legend()
    if annotate_scores:
        props = dict(boxstyle='round', facecolor='wheat', alpha=0.7)
        textstr = ('F1: {}\nprecision: {}\nrecall: {}\nthreshold: {}'
                   .format(f1_score['F1'], f1_score['precision'],
                           f1_score['recall'], threshold))
        ax2.text(0.15, 0.95, textstr, transform=ax2.transAxes,
                verticalalignment='top', bbox=props)
    ax1.set_ylabel('signals')
    ax2.set_ylabel(density_column)
    ax1.legend(loc=4)
    ax2.legend()
    if use_plotly:
        py.iplot_mpl(f)
        fig = tls.mpl_to_plotly(f, resize=True)
        fig['layout'].update(showlegend=True)
        fig.strip_style()
        py.iplot(fig)
    return


def train_amgng(signal, dimensions, buffer_len, ma_window,
                ma_recalc_delay, increase_start, decrease_stop,
                modify_time_domain=False, modify_freq_domain=False,
                alpha=0.5, beta=0.75, gamma=None,
                delta=0.5, theta=None, eta=0.9995, lmbda=None,
                e_w=0.05, e_n=0.0006, ddof=1):
    modification_len = decrease_stop - increase_start
    if modify_freq_domain:
        trans = np.fft.fftn(signal.iloc[increase_start:decrease_stop])
        noise = trans + 2*np.random.randn(modification_len, dimensions)
        itrans = np.fft.ifftn(noise).real
        signal.iloc[increase_start:decrease_stop] = itrans
    if modify_time_domain:
        signal.iloc[increase_start:decrease_stop] += np.random.randn(modification_len, dimensions)

    if theta is None:
        theta = buffer_len
    if gamma is None:
        gamma = theta//2
    if lmbda is None:
        lmbda = theta*6
    amgng = AMGNG(comparison_function=compare_models,
                  buffer_len=buffer_len, dimensions=dimensions,
                  prest_gamma=gamma, prest_lmbda=lmbda,
                  prest_theta=theta, pst_gamma=gamma,
                  pst_lmbda=lmbda, pst_theta=theta,
                  ma_window_len=ma_window, ma_recalc_delay=ma_recalc_delay, ddof=ddof,
                  prest_alpha=alpha, prest_beta=beta, prest_delta=delta,
                  prest_eta=eta, prest_e_w=e_w, prest_e_n=e_n,
                  pst_alpha=alpha, pst_beta=beta, pst_delta=delta,
                  pst_eta=eta, pst_e_w=e_w, pst_e_n=e_n)
    scores = np.zeros(len(signal))
    pscores = np.zeros(len(signal))
    print('Processing {} rows'.format(len(signal)))
    for t, xt in enumerate(signal.values):
        if t % (len(signal)//100) == 0:
            print('{}% done. Sample datapoint: {}'
                  .format(t / (len(signal)//100), xt))
        scores[t], pscores[t] = amgng.time_step(xt)

    signal['anomaly_density'] = pscores
    signal['anomaly_score'] = scores
    signal['Annotation'] = [''] * len(signal)
    signal['Annotation'].iloc[increase_start:decrease_stop] = 'A'
    return signal


def train_cla(name, cwd, file_name, signal, buffer_len,
              increase_start, decrease_stop):

    full_path = os.path.join(cwd, file_name)
    create_cla_input_file(full_path, signal)
    fields = signal.columns.tolist()
    cla_model = {}
    _, csv_reader, input_handler = open_input_file(full_path)
    for p in fields:
        swarm_desc = fill_swarm_description(name, 
                                            full_path, buffer_len,
                                            fields, p)
        model_params = swarm(cwd=cwd, input_file=file_name,
                             swarm_description=swarm_desc)
        model = create_model(params=model_params, predictedField=p)
        model_out_file = os.path.join(cwd, '{}_{}'.format(p, file_name))
        shifter, output_handler = prepare_run(fields=fields,
                                              predicted_field=p,
                                              plot=False,
                                              output_name=model_out_file,
                                              index_name='step')
        cla_model[p] = {'model': model, 'shifter': shifter,
                        'output_handler': output_handler,
                        'model_out_file': model_out_file}

    for i, row in enumerate(csv_reader):
        for p in fields:
            start = datetime.now()
            process_row(row=row, fields=fields, predicted_field=p,
                        model=cla_model[p]['model'],
                        shifter=cla_model[p]['shifter'],
                        output_handler=cla_model[p]['output_handler'],
                        counter=i,
                        datetime_index=False)

    input_handler.close()
    n_rows = i
    for i, p in enumerate(fields):
        cla_model[p]['output_handler'].close()
        df = pd.read_csv(cla_model[p]['model_out_file'], parse_dates=True,
                         index_col='step')
        if i == 0:
            cla_df = df
        else:
            cla_df.anomaly_likelihood.iloc[:n_rows] += df.anomaly_likelihood.iloc[:n_rows]
            cla_df.anomaly_score.iloc[:n_rows] += df.anomaly_score.iloc[:n_rows]
    cla_df.anomaly_likelihood /= len(fields)
    cla_df.anomaly_score /= len(fields)
    cla_df['Annotation'] = [''] * len(cla_df)
    cla_df['Annotation'].iloc[increase_start:decrease_stop] = 'A'
    return cla_df


def get_f1_curve(signal, density_column, regex, annotation_column):
    f1 = partial(utils.f1_score, df=signal, col=density_column,
                                 match=regex, annotation_column=annotation_column,
                                 scalar=True)
    max_score = signal[density_column].max()
    x = np.linspace(0, max_score, num=max_score*200)
    y = [f1(t) for t in x]
    
    return x, y


def get_scores_curve(signal, density_column, regex, annotation_column):
    f1 = partial(utils.f1_score, df=signal, col=density_column,
                                 match=regex, annotation_column=annotation_column,
                                 scalar='F1')
    precision = partial(utils.f1_score, df=signal, col=density_column,
                                        match=regex, annotation_column=annotation_column,
                                        scalar='precision')
    recall = partial(utils.f1_score, df=signal, col=density_column,
                                     match=regex, annotation_column=annotation_column,
                                     scalar='recall')
    max_score = signal[density_column].max()
    x = np.linspace(0, max_score, num=max_score*200)
    f = [f1(t) for t in x]
    p = [precision(t) for t in x]
    r = [recall(t) for t in x]
    
    return x, f, p, r


def plot_scores(xs, f1, precision, recall):
    f = figure()
    ax1 = subplot(2, 2, 1)
    ax1.plot(xs, f1)
    ax1.set_xlabel('anomaly_density threshold')
    ax1.set_ylabel('F1-Score')
    ax2 = subplot(2, 2, 2)
    ax2.plot(xs, precision)
    ax2.set_xlabel('anomaly_density threshold')
    ax2.set_ylabel('Precision')
    ax3 = subplot(2, 2, 3)
    ax3.plot(xs, recall)
    ax3.set_xlabel('anomaly_density threshold')
    ax3.set_ylabel('Recall')


def find_model_w_best_window(signal, dimensions, buffer_len, ma_recalc_delay,
                             increase_start, decrease_stop,
                             max_window=None):
    ma_scores = []
    if max_window is None:
        max_window = len(signal)//2
    for i in range(1, (max_window//50) + 1):
        ma_window = 50 * i
        try:
            np.random.seed(0)
            tmp_signal = signal.copy()
            tmp_signal = train_amgng(signal=tmp_signal, dimensions=dimensions,
                                 buffer_len=buffer_len, ma_window=ma_window,
                                 ma_recalc_delay=ma_recalc_delay,
                                 increase_start=increase_start,
                                 decrease_stop=decrease_stop,
                                 modify_time_domain=modify_time_domain,
                                 modify_freq_domain=modify_freq_domain)
        except Exception as e:
            print(signal)
            raise e
        xs, scores = get_f1_curve(tmp_signal, density_column='anomaly_density',
                                  regex='A', annotation_column='Annotation')
        max_score = max(scores)
        ma_scores.append((max_score, ma_window, tmp_signal))
    best_model = max(ma_scores)
    max_score = best_model[0]
    ma_max_score = best_model[1]
    signal_max_score = best_model[2]
    return max_score, ma_max_score, signal_max_score


def create_cla_input_file(file_name, signal):
    with open(file_name, 'wb') as file_out:
        writer = csv.writer(file_out)
        header = ['step'] + signal.columns.tolist()
        writer.writerow(header)
        types = ['int'] + ['float' for _ in header[1:]]
        writer.writerow(types)
        flags = ['' for _ in header]
        writer.writerow(flags)
        writer.writerows(signal.itertuples())
    return


def fill_swarm_description(name, file_path, buffer_len,
                           fields, predicted_field, swarm_size='small'):
    swarm_desc = {
        'includedFields': [
            {
                'fieldName': 'step',
                'fieldType': 'int',
            },
        ],
        'streamDef': {
            'info': name,
            'version': 1,
            'streams': [
                {
                    'info': name,
                    'source': file_path,
                    'columns': ['*']
                }
            ]
        },
        'inferenceType': 'TemporalAnomaly',
        'inferenceArgs': {
            'predictionSteps': [1],
            'predictedField': predicted_field
        },
        'iterationCount': buffer_len,
        'swarmSize': swarm_size
    }
    for field in fields:
        swarm_desc['includedFields'].append(
                {
                    'fieldName': field,
                    'fieldType': 'float',
                })
    return swarm_desc

def get_best_ma_window(signal, step, max_ma_window=None):
    print(signal.head())
    if max_ma_window is None:
        max_ma_window = len(signal)//2
    best_params = None
    best_f1 = 0
    for ma_len in xrange(step, max_ma_window + step, step):
        # stats[0] are the means; stats[1] are the stds
        scores = signal.anomaly_score.iloc[ma_len:]
        stats_params = np.asarray([(bn.nanmean(scores[i:i+ma_len]), bn.nanstd(scores[i:i+ma_len], ddof=1))
                                   for i in xrange(len(scores))]).T
        densities = np.zeros(len(signal))
        densities[ma_len:] += norm.cdf((scores - stats_params[0])/stats_params[1])
        signal['temp_anomaly_density'] = densities
        xs, f1, precision, recall = get_scores_curve(signal, 'temp_anomaly_density', 'A', 'Annotation')
        max_f1 = max(f1)
        if max_f1 >= best_f1:
            best_f1 = max_f1
            best_params = {'ma_window': ma_len, 'xs': xs, 'f1': f1, 'precision': precision, 'recall': recall,
                           'anomaly_density': signal.temp_anomaly_density}
    return best_params


def print_f1_stats(signal, density_column, regex, annotation_column):
    xs, f1, precision, recall = get_scores_curve(signal=signal, density_column='anomaly_likelihood',
                                                 regex='A', annotation_column='Annotation')
    max_f1_index = np.argmax(f1)
    max_threshld = xs[max_f1_index]
    f1_stats = utils.f1_score(max_threshld, signal, 'anomaly_likelihood', 'A')
    print('Real: {real};\n'
          'Threshold: {threshold};\n'
          'Detected: {detected};\n'
          'True Positives: {true_positives};\n'
          'False Positives: {false_positives};\n'
          'False Negatives: {false_negatives};\n'
          'Precision: {precision};\n'
          'Recall: {recall};\n'
          'F1: {F1}'
          .format(**f1_stats))
    return


def print_shift_stats(signal, shift_position):
    exprmnt_stats = {
        'start_mean_score_after_100': bn.nanmean(signal.anomaly_score[100:shift_position]),
        'start_std_score_after_100': bn.nanstd(signal.anomaly_score[100:shift_position], ddof=1),
        'start_mean_score_after_1000': bn.nanmean(signal.anomaly_score[1000:shift_position]),
        'start_std_score_after_1000': bn.nanstd(signal.anomaly_score[1000:shift_position], ddof=1),
        'start_mean_score_after_10000': bn.nanmean(signal.anomaly_score[10000:shift_position]),
        'start_std_score_after_10000': bn.nanstd(signal.anomaly_score[10000:shift_position], ddof=1),
        'mean_score_before': bn.nanmean(signal.anomaly_score[:shift_position]),
        'std_score_before': bn.nanstd(signal.anomaly_score[:shift_position], ddof=1),
        'mean_score_after': bn.nanmean(signal.anomaly_score[shift_position:]),
        'std_score_after': bn.nanstd(signal.anomaly_score[shift_position:], ddof=1),
        'mean_score_after_100': bn.nanmean(signal.anomaly_score[shift_position+100:]),
        'std_score_after_100': bn.nanstd(signal.anomaly_score[shift_position+100:], ddof=1),
        'mean_score_after_1000': bn.nanmean(signal.anomaly_score[shift_position+1000:]),
        'std_score_after_1000': bn.nanstd(signal.anomaly_score[shift_position+1000:], ddof=1),
        'mean_score_after_10000': bn.nanmean(signal.anomaly_score[shift_position+10000:]),
        'std_score_after_10000': bn.nanstd(signal.anomaly_score[shift_position+10000:], ddof=1),
    }
    print(
          'Score mean before shift: {mean_score_before};\n'
          'Score std before shift: {std_score_before};\n'
          'Score mean after shift: {mean_score_after};\n'
          'Score std after shift: {std_score_after};\n'
          'Start Score mean after shift 100: {start_mean_score_after_100};\n'
          'Start Score std after shift 100: {start_std_score_after_100};\n'
          'Start Score mean after shift 1000: {start_mean_score_after_1000};\n'
          'Start Score std after shift 1000: {start_std_score_after_1000};\n'
          'Start Score mean after shift 10000: {start_mean_score_after_10000};\n'
          'Start Score std after shift 10000: {start_std_score_after_10000};\n'
          'Score mean after shift 100: {mean_score_after_100};\n'
          'Score std after shift 100: {std_score_after_100};\n'
          'Score mean after shift 1000: {mean_score_after_1000};\n'
          'Score std after shift 1000: {std_score_after_1000};\n'
          'Score mean after shift 10000: {mean_score_after_10000};\n'
          'Score std after shift 10000: {std_score_after_10000}'
          .format(**exprmnt_stats))
```

```python
def run_amgng_mackey_glass_shift(x):
    dimensions = 4
    buffer_len = 200
    sample_len = 150000
    increase_start = 74600
    increase_stop = 75400
    decrease_start = 75600
    decrease_stop = 76400
    gamma = 88
    e_w = 0.4
    e_n = 0.05
    ma_window = 38000
    ma_recalc_delay = 1
    ddof = 1
    modification_len = decrease_stop - increase_start
    signal, annotations = signal_generation.generate_mackey_glass(increase_start, increase_stop,
                                                                  decrease_start, decrease_stop,
                                                                  n_samples=dimensions,
                                                                  sample_len=sample_len,
                                                                  increases=[0, 0, 0, 0],
                                                                  seed=0)
    signal.iloc[:,1] *= -1
    signal.iloc[:,2] *= 2
    signal.iloc[:,3] *= -2
    signal.iloc[increase_start:, 0] += 5
    signal.iloc[increase_start:, 1] -= 5
    signal.iloc[increase_start:, 2] += 10
    signal.iloc[increase_start:, 3] -= 10

    with Capturing() as output:
        signal = train_amgng(signal, dimensions, buffer_len, ma_window,
                             ma_recalc_delay, increase_start, decrease_stop,
                             gamma=gamma, e_w=e_w, e_n=e_n, ddof=ddof)


    ret_val = {
        'increase_start': increase_start,
        'decrease_stop': decrease_stop,
        'buffer_len': buffer_len,
        'ma_window': ma_window,
        'signal': signal,
    }
    return ret_val

amgng_mackey_glass_shift = ldview.map_async(run_amgng_mackey_glass_shift, [0])
```
